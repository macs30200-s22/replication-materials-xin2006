```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import csv
```

```python
crime_df = pd.read_excel(r'Crimes_2018.1_to_Present.xlsx')
```

```python
crime_df.head()
```

```python
crime_df.columns
```

```python
crime=crime_df.drop(['ID','Case Number','Block', 'IUCR','Description','Location Description','Arrest','Beat','Ward','FBI Code',
               'X Coordinate','Y Coordinate','Updated On', 
               'Latitude', 'Longitude','Location'],axis=1)
```

```python
crime['Month'] = crime_df.Date.str.extract(pat='(^[0-9][0-9])', expand=False)
crime.Month
```

```python
crime['Year']=crime['Year'].map(lambda x:str(x))
```

```python
crime['Month_Year']=crime['Month'].str.cat(crime['Year'],sep="/")
```

```python
crime.head()
```

```python
crime=crime.drop(['Date','Year','Month'],axis=1)
```

```python
crime[0:10]
```

```python
len(crime)
```

```python
type_set=set()
for i in crime['Primary Type']:
    type_set.add(i)
type_set
```

```python
# Count for domestic crime
count_domestic={}
for i in range(len(crime)):
    if crime['Domestic'][i]==True:
        if crime['Month_Year'][i] not in count_domestic:
            count_domestic[crime['Month_Year'][i]]=0
        count_domestic[crime['Month_Year'][i]]+=1
    if crime['Month_Year'][i]==0:
        print(i)

count_domestic   
```

```python
# Count for domestic crime
count_indomestic={}
for i in range(len(crime)):
    if crime['Domestic'][i]==False:
        if crime['Month_Year'][i] not in count_indomestic:
            count_indomestic[crime['Month_Year'][i]]=0
        count_indomestic[crime['Month_Year'][i]]+=1
    if crime['Month_Year'][i]==0:
        print(i)

count_indomestic   
```

```python
for i in count_domestic:
    with open('type_domestic.csv','a') as f:
        wf = csv.writer(f)
        wf.writerow([i,count_domestic[i],count_indomestic[i]])
```
